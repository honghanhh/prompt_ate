# TERM LENGTHS

```python
def computeTermEvalMetrics(extracted_terms, gold_df):
    # make lower case cause gold standard is lower case
    extracted_terms = [item.lower() for item in extracted_terms]
    extracted_terms = set([x for x in extracted_terms if x != ''])
    gold_set = [item.lower() for item in gold_df]
    gold_set =  set([x for x in gold_set if x != ''])
    true_pos = extracted_terms.intersection(gold_set)
    recall = round(len(true_pos)*100/len(gold_set),1)
    precision = round(len(true_pos)*100/len(extracted_terms),1)
    fscore = round(2*(precision*recall)/(precision+recall),1) if precision+recall != 0 else 0
    print(str(len(extracted_terms)) +' & ' +  str(len(gold_set)) +' & ' + str(len(true_pos)) +' & ' +  str(precision)+' & ' +  str(recall)+' & ' +  str(fscore))
    return len(extracted_terms), len(gold_set), len(true_pos), precision, recall, fscore

def extract_between_markers(input_string):
    pattern = r'@@(.*?)##'
    extracted = re.findall(pattern, input_string)
    return extracted

def extract_entities(words, labels):
    entities = []
    current_entity = ''
    current_label = ''

    for word, label in zip(words, labels):
        if label.startswith('B'):
            if current_entity:
                entities.append(current_entity.strip())
            current_entity = word
            current_label = label[2:]
        elif label.startswith('I') and current_label == label[2:]:
            current_entity += ' ' + word
        else:
            if current_entity:
                entities.append(current_entity.strip())
                current_entity = ''
                current_label = ''

    if current_entity:
        entities.append(current_entity.strip())

    return entities

def filter_text(text):
    allowed_chars = {'B', 'I', 'O', ' '}
    filtered_text = ''.join(c for c in text if c in allowed_chars)
    return filtered_text

def pred_gt(preds, gts):
    preds_1 = [str(x).lower() for x in preds if len(str(x).split()) == 1 ]
    gts_1 = [str(x).lower()  for x in gts if len(str(x).split()) == 1]
    preds_2 = [str(x).lower()  for x in preds if len(x.split()) == 2]
    gts_2 = [str(x).lower() for x in gts  if len(str(x).split()) == 2]
    preds_3 = [str(x).lower() for x in preds if len(str(x).split()) == 3]
    gts_3 = [str(x).lower() for x in gts if len(str(x).split()) == 3]
    preds_4 = [str(x).lower() for x in preds if len(str(x).split()) == 4]
    gts_4 = [str(x).lower() for x in gts if len(str(x).split()) == 4 ]
    preds_5 = [str(x).lower()  for x in preds if len(str(x).split()) >= 5]
    gts_5 = [str(x).lower() for x in gts if len(str(x).split()) >= 5 ]
    computeTermEvalMetrics(preds_1, gts_1)
    computeTermEvalMetrics(preds_2, gts_2)
    computeTermEvalMetrics(preds_3, gts_3)
    computeTermEvalMetrics(preds_4, gts_4)
    computeTermEvalMetrics(preds_5, gts_5)
```

```python
en_ann_output1 = []
for x, y in list(zip(en_htfl_preds.words, en_htfl_preds.en_ann_output1)):
    words = eval(x)
    labels = y[:len(words)*2]
    if labels.startswith("'I"):
        labels = 'O '*len(words)
    labels =  filter_text(labels).split(' ')
    en_ann_output1.append(labels)
en_htfl_preds['en_ann_output1_processed'] = en_ann_output1

en_ann_output1_processed = []
for x, y in zip(en_htfl_preds.words, en_htfl_preds.en_ann_output1_processed):
    terms =  extract_entities(eval(x), y)
    en_ann_output1_processed.extend(terms)

    
en_ann_output2 = []
for x in en_htfl_preds['en_ann_output2']:
    if x.startswith('[') and x.endswith(']') and x.count('[') == 1: 
        if "Jehovah's Witness" in x:
            en_ann_output2.extend(["18-year-old", "Jehovah's Witness", "sickle cell disease", "life-threatening anemia"])
        else:
            en_ann_output2.extend(eval(x))
            
en_htfl_preds['en_ann_output3_processed'] = [extract_between_markers(x) for x in en_htfl_preds['en_ann_output3']]
en_ann_output3_processed = []
for x in en_htfl_preds['en_ann_output3_processed']:
    en_ann_output3_processed.extend(x)
```
